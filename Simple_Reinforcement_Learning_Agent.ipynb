```python
import random
import pandas as pd
```

## Defining Environment

```python
class Environment:
  def __init__(self):
    self.steps_left=100

  def get_observation(self):
    return [0.0,0.0,0.0,0.0]

  def get_actions(self):
    return [0,1]

  def is_done(self):
    return self.steps_left==0

  def action(self,action):
    if self.is_done():
      print("Game Over")
    self.steps_left-=1
    return random.random() #Return random number between 0.0 and 1.0
```

## Defining Agent

```python
class Agent:
  def __init__(self):
    self.total_reward=0.0

  def steps(self,env):
    current_obs=env.get_observation()
    actions=env.get_actions()
    reward=env.action(random.choice(actions)) #Return a random element from a list [0,1]
    self.total_reward+=reward

#Instantiate the classes using constructors
env=Environment()
agent=Agent()

while not env.is_done():
  agent.steps(env)

print("Total reward got: %d "% agent.total_reward)
```

```python
steps=[]
rewards=[]
for i in range(5,51,2):
  env=Environment()
  agent=Agent()

  while not env.is_done():
    agent.steps(env)

  steps.append(i)
  rewards.append(agent.total_reward)

df=pd.DataFrame({"steps":steps,"rewards":rewards})
df
```
